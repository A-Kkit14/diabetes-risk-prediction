```python
# kiểm tra dữ liệu continuous có chứa bao nhiêu các outlier
numerical_columns = ['PhysHlth', 'MentHlth', 'BMI']

dict_outlier = {}
for col_name in numerical_columns:
    series_data = data[col_name]
    q1 = series_data.quantile(0.25)
    q3 = series_data.quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    outliers = series_data[(series_data < lower_limit) | (series_data > upper_limit)]
    dict_outlier[col_name] = len(outliers)

dict_outlier
```

# Tiền xử lý dữ liệu

```python
# chỉnh lại kiểu dữ liệu cho một số biến nhị phân và biến phân loại
binary_columns = []
for column_name in data:
    unique_values = data[column_name].unique()
    if unique_values.shape[0] == 2:
        binary_columns.append(column_name)

numerical_columns = ['PhysHlth', 'MentHlth', 'BMI']

ordinal_columns = set(data.columns) - set(numerical_columns) - set(binary_columns)

binary_ordinal_columns = binary_columns + list(ordinal_columns)

data[binary_ordinal_columns] = data[binary_ordinal_columns].astype(int)
```

```python
path = os.path.join('/kaggle/working', 'data_clean.csv')
data_clean = data.copy()

data_clean.to_csv(path, index=False)
```
